```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df = df[['Pclass','Sex','Age','Fare','Survived']]
df.isnull().sum()
```

```python
df = df[df['Age'].notna()]
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
inputs = df.drop(columns='Survived')
```

```python
inputs
```

```python
target = df['Survived']
target.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
inputs['Sex'] = le.fit_transform(inputs.Sex)
inputs.head()
```

```python
inputs[['Age','Fare']] = inputs[['Age','Fare']].astype(int)
```

```python
model.fit(inputs,target)
```

```python
model.score(inputs,target)
```

```python
model.predict([[3,1,22,7]])
```
