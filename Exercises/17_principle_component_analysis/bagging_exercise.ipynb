```python
import pandas as pd
df = pd.read_csv("heart.csv")
df.head()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
```

```python
df.describe()
```

```python
df.shape
```

```python
numeric_df = df.select_dtypes(include=[np.number])
```

```python
numeric_df
```

```python
filtered_df = df[(np.abs(zscore(numeric_df)) <= 3).all(axis=1)]
```

```python
filtered_df.shape
```

```python
filtered_df.head()
```

```python
filtered_df['ChestPainType'].unique()
```

```python
filtered_df["ST_Slope"].unique()
```

```python
filtered_df['ST_Slope'] = filtered_df['ST_Slope'].map({
    'Up' : 0,
    'Flat' : 1,
    'Down' : 2
})
```

```python
filtered_df = filtered_df.copy()
```

```python
filtered_df
```

```python
encoded_columns = ['Sex','ChestPainType','RestingECG','ExerciseAngina']
encoded_df = pd.get_dummies(filtered_df[encoded_columns],drop_first=True)
encoded_df.head()
```

```python
concated_df = pd.concat([filtered_df[['Age','RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak','ST_Slope']],encoded_df],axis=1)
concated_df.head()
```

```python
concated_df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(concated_df)
scaled_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_df,filtered_df.HeartDisease)
```

```python
X_train
```

```python
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(X_train,y_train)
svm_model.score(X_test, y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
clf_svc = BaggingClassifier(SVC(),n_estimators=100,max_samples=0.90,oob_score=True)
```

```python
clf_svc.fit(X_train,y_train)
```

```python
clf_svc.oob_score_
```

```python
clf_svc.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
dt_model.score(X_test, y_test)
```

```python
clf_dt = BaggingClassifier(DecisionTreeClassifier(),n_estimators=100,max_samples=0.90,oob_score=True)
```

```python
clf_dt.fit(X_train,y_train)
```

```python
clf_dt.oob_score_
```

```python
clf_dt.score(X_test,y_test)
```

### Performance of DecisionTreeClassifier improves more than SVC because Trees are high variance models -> they benefit a lot from bagging. Bagging reduces variance, making predictions more stable and accurate.

### Whereas, SVM's are low-variance, high bias models -> They are already quite stable; bagging has less effect.
