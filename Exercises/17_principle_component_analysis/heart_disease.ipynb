```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(df[['Age','RestingBP','Cholesterol','Oldpeak']]))
```

```python
threshold_z = 3
outlier_indices = np.unique(np.where(z > threshold_z)[0] )
outlier_indices
```

```python
df_no_outlier = df.drop(outlier_indices)
```

```python
df_no_outlier.shape
```

```python
df_no_outlier.sample(50)
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
df_encoded = pd.get_dummies(df_no_outlier,columns=['ChestPainType', 'RestingECG', 'ST_Slope'],drop_first=True)
df_encoded['Sex'] = le.fit_transform(df_encoded['Sex'])
df_encoded['ExerciseAngina'] = le.fit_transform(df_encoded['ExerciseAngina'])

df_encoded.head()
```

```python
df_encoded.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_encoded.drop('HeartDisease',axis=1))
scaled_data
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_data, df_encoded.HeartDisease,test_size=0.2)
```

```python
from sklearn.model_selection im
```

```python
from sklearn import svm
model = svm.SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_log = LogisticRegression()
model_log.fit(X_train, y_train)
model_log.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
model_rf.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(x_pca, df_encoded.HeartDisease, test_size=0.2)
```

```python
X_train_pca.shape
```

```python
model_with_pca = svm.SVC()
```

```python
model_with_pca.fit(X_train_pca, y_train_pca)
model_with_pca.score(X_test_pca, y_test_pca)
```
