```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("test_scores.csv")
df
```

```python
x = np.array(df.math).reshape(-1,1)
y = np.array(df.cs)
```

```python
def gradient(x,y):
    curr_m = curr_b = 0
    iterations = 1000
    learning_rate = 0.0002
    n = len(x)
    prev_cost = float('inf')
    for i in range(iterations):
        predicted_y = curr_m * x + curr_b
        curr_cost = (1/n) * sum((y - predicted_y) ** 2)
        if math.isclose(prev_cost, curr_cost, rel_tol=1e-20):
             print(f"Converged at iteration {i}")
             break
           
        prev_cost = curr_cost
        md = -(2/n) * sum(x * (y-predicted_y))
        bd = -(2/n) * sum(y-predicted_y)
        curr_m = curr_m - learning_rate * md
        curr_b = curr_b - learning_rate * bd
        print("m {:.4f} b {:.4f} cost {:.6f} iterations {}".format(curr_m,curr_b,curr_cost,i))

        

            
       
        
        
        
    
```

```python
gradient(x,y)
```
